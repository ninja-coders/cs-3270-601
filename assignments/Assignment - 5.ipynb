Assignment 5 - Simple Relational Algebra Interpreter
============================

For this assignment you will be required to write a simple relational algebra
interpreter. A relation is nothing more than a table of fixed-size data rows,
which you can store as a set of tuples. All the required functionality can
reside in a single Table class. Here is the `interface` you should implement:    

    class Table:
        def __init__(self,name='',fields=tuple(),tups=None): …

        # Relational operations:
        def select(self,field,val): …
        def project(self,*fields): …
        @staticmethod
        def join(tab1,tab2): …
        def insert(self,*tup): …
        def remove(self,field,val): …

        # Serialization and text backup
        def store(self): …
        @staticmethod
        def restore(fname): …
        @staticmethod
        def read(fname): …
        def write(self,fname): …

## Interface details

A Table has 3 data attributes: 
* a name
* a group of fields (which defines the columns)
* a set of tuples (the rows). 

**Note:** Use the “____” prefix trick for these attributes so they are
not inadvertently changed by users and provide properties for them as follows:   
* name:read/write
* fields: read only
* tuples: read only

The following methods return a new Table as a result: 
* select
* project
* join
* restore
* read. 

The first three should name the returned table “result”. The last two read
their names and other data from files. The other methods do not return anything.

To illustrate, suppose the file parts.txt contains the following text:

    parts
    pno,pname,color,weight,city
    p1,Nut,Red,12,London
    p2,Bolt,Green,17,Paris
    p3,Screw,Blue,17,Rome
    p4,Screw,Red,14,London
    p5,Cam,Blue,12,Paris
    p6,Cog,Red,19,London
    
The first line is the table name, and the second line contains the field (column) names.
The rest of the lines represent the tuples (rows) of the table.  

### Read Method

This is a demonstration of the read of a table:  

    >>> parts = Table.read(‘parts.txt’)
    >>> print(parts)
    
    parts('pno', 'pname', 'color', 'weight', 'city')
    =====
    ('p1', 'Nut', 'Red', '12', 'London')
    ('p4', 'Screw', 'Red', '14', 'London')
    ('p6', 'Cog', 'Red', '19', 'London')
    ('p3', 'Screw', 'Blue', '17', 'Rome')
    ('p5', 'Cam', 'Blue', '12', 'Paris')
    ('p2', 'Bolt', 'Green', '17', 'Paris')
    
The output format is defined here:  
* The first line is the table name followed by the tuple of field names.  
* Then a line which underlines the table name is printed
* After that each tuple is printed on a line by itself. 

**Note:** For the above code to work you will need to overwrite the `__str__`
method of the table that returns a string with the supplied format.   

**Note:** Also to watch for empty lines at the end of the file that should
be ignored.

### Store and Restore

The store and restore methods merely use pickle to serialize/deserialize a Table
object.   

Store writes to a file named `<name>.db`, where name is the internal name of
the table.   

**Hint: (Lookup pickle.dump and pickle.load in the Python documentation.)**   

### Select Method

The select method takes a field name and a value as parameters and returns a result
table with each row having the value of the specified field matching the passed in value.   

For example:   

    >>> print(parts.select(‘pno’,'p1'))
    
    result('pno', 'pname', 'color', 'weight', 'city')
    ======
    ('p1', 'Nut', 'Red', '12', 'London')
    
### Project Method

The project method takes one or more field names and returns a result table with only
the associated columns from each row:   

    >>> print parts.project('pname','color')
    
    result('pname', 'color')
    ======
    ('Bolt', 'Green')
    ('Nut', 'Red')
    ('Screw', 'Blue')
    ('Screw', 'Red')
    ('Cam', 'Blue')
    ('Cog', 'Red')

### Insert Method

The insert method is a no-brainer, except make sure that you only attempt to add tuples
to a table that have the correct number of fields **(we want the tables to be perfectly
rectangular).**  

And of course, tables allow no duplicate tuples. 

### Remove Method

The remove method removes all tuples from a table where the indicated field matches 
the passed-in value.
